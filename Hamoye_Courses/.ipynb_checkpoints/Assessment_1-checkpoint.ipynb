## Q3 - What is the total number and percentage of missing data in 2016 to 2 decimal places?

```python
df.isnull().sum()
```

```python
percent_missing = df.isnull().sum() * 100 / len(df)
missing_value_df = pd.DataFrame({'column_name': df.columns,
                                 'percent_missing': percent_missing})

missing_value_df
```

## Q4 - Which year had the highest correlation with ‘Element Code’? 

```python
df.corr()
```

```python
#ploting the heatmap for correlation
corr_matrix = df.corr()
fig, ax = plt.subplots(figsize=(12,8))
ax = sns.heatmap(corr_matrix,
                annot=True,
                linewidths=0.5,
                fmt=".2f",
                cmap="YlGnBu");
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5);
```

## Q5 - What year has the highest sum of Import Quantity?

```python
df.head(2)
```

```python
df.groupby('Element')['Y2014', 'Y2015','Y2016', 'Y2017','Y2018'].sum()
```

```python
x = df.groupby('Element')[['Y2014', 'Y2015','Y2016', 'Y2017','Y2018']].sum()
y = x.sort_values(by=['Element'],ascending=True)

y.plot(kind='bar', figsize=(12,8))
plt.show()
```
