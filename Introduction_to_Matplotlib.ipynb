## Anatomy of a Matplotlib fig

```python
# A matplotlib workflow

# 0. Import and get matplotlib ready
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10))

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Sample Simple Plot", xlabel="x-axis", ylabel="y-axis")

# 5. Save & show
# fig.savefig("Files/Matplotlib/simple-plot.png")
```

## Making Figures with Numpy Arrays

We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
# Create some data
x = np.linspace(0,10,100)
x
```

```python
# Plot the data
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Still scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond butter": 10, 
                     "Peanut butter": 8, 
                     "Cashew butter": 12}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title = "Nut Butter Store",
      ylabel = "Price (£)")
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond butter": 10, 
                     "Peanut butter": 8, 
                     "Cashew butter": 12}

fig, ax = plt.subplots()

#Bar graph
x.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title = "Nut Butter Store",
      ylabel = "Price (£)"); # add a semi-colon to remove the text
```

```python
# Horizonal bargraph
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Histogram
x = np.random.randn(1000)

fig, ax = plt.subplots()
ax.hist(x);
```

```python
x = np.random.random(1000) # random data from random distribution

fig, ax = plt.subplots()
ax.hist(x);
```

# Subplots

```python
# Subplot option 1

fig,((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5))
# plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))

# Plot to each different index
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

## Plotting from pandas DataFrames

```python
# Make a Dataframe
car_sales = pd.read_csv("Files/Pandas/car-sales.csv")
car_sales
```

```python
 # Start with some dummy data
ts = pd.Series(np.random.randn(1000),
               index=pd.date_range('1/1/2020', periods=1000))
ts
```

```python
# What does cumsum() do? Return cumulative sum over a Dataframe
# Column = [1,2,3,4,5]
# Column.cumsum() = [1, 3(1+2), 6(3+3), 10(6+4), 15(10+5)]
ts.cumsum()
```

```python
ts.cumsum().plot();
```

```python
# Remove price column symbols
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '')
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales["Sales Day"] = pd.date_range("20/11/2021", periods=len(car_sales))
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
# Let's plot the total sales
car_sales.plot(x="Sales Day", y="Total Sales");
```

```python
# Doesn't work
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter");
```

```python
# Convert Price to int
car_sales["Price"] = car_sales["Price"].astype(int)

# Plot scatter plot with price column as numeric
car_sales.plot(x="Odometer (KM)", y="Price", kind='scatter');
```
