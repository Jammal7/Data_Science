## 3. How are you going to increase the sales for the products that least bought also showing correlations between variables


```python
## grouping product by their "Quantity","Unit price","Shipping Cost","Discount" and checking their correlation
product_quant = pd.DataFrame(data.groupby("Product Name")["Quantity","Market","Unit price","Shipping Cost","Discount"].sum())
product_quant["average_discount"] = product_quant["Discount"]/product_quant["Quantity"]
product_quant["average_shippingcost"] = product_quant["Shipping Cost"]/product_quant["Quantity"]
product_quant.head(10)
```

```python
product_quant.corr()#correlation
```

```python
plt.figure(figsize = (16,10))
#creating mask for the heat map
mask = np.zeros_like(product_quant.corr())
upper_part = np.triu_indices_from(mask)
mask[upper_part] = 1
plt.title("Relationship between Quantity Unit price Shipping Cost Discount average discount and average shipping cost", fontsize=20)
plt.xlabel("variables",fontsize=20)
sns.heatmap(product_quant.corr(),mask = mask, annot =True, annot_kws={"size":14},vmin = 0,vmax = 1)
```

```python
# Creating a new dataframe consisting of top 15 performing product
#nlargest give the highest number or lrgest number
Top_products = pd.DataFrame(data.groupby("Product Name")["Unit price",'Market',"Shipping Cost","Discount","Quantity"].sum().nlargest(10,columns = "Quantity"))
Top_products
```

```python
mask = np.zeros_like(Top_products.corr())
upper_part = np.triu_indices_from(mask)
mask[upper_part] = 1
sns.heatmap(Top_products.corr(),mask = mask, annot =True, annot_kws={"size":14})
```

```python
Top_products.plot(kind="bar",x="Quantity")
```

```python
poor_products = pd.DataFrame(data.groupby("Product Name")["Quantity","Unit price","Shipping Cost","Discount",].sum().nsmallest(10,columns="Quantity"))
poor_products
```

```python
#Generating Sales of 2013
#picking 20l3 ordered years
yr_2013_data = pd.DataFrame(data.loc[data["Order_Year"] == 2013])
trend_2013 = pd.DataFrame(yr_2013_data.groupby("Order_Month")["Sales"].sum()).reset_index()
#convert ordered month back from month abrv in string to date.time month
trend_2013["Order_Month"] = pd.to_datetime(trend_2013["Order_Month"], format='%b', errors='coerce').dt.month
#sort out based on date.time month
trend_2013 = trend_2013.sort_values(by="Order_Month")
trend_2013
```
